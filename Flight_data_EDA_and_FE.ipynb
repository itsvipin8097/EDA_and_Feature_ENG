```python
# Now we are going to convert date_of_journey column into numeric
# Now we can see date month and year is seprated in list
df['Date_of_Journey'].str.split("/")
```

```python
df['Date']=df['Date_of_Journey'].str.split("/").str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2) #now we can see we have three seprate column for dates
```

```python
df['Date'].dtype # this is still in string type lets convert this into int
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
# Now we can drop column date of journey

df.drop(['Date_of_Journey'],axis=1,inplace=True)
```

```python
df.head()
```

```python
# lets do for total_stops Now we can see there are 5 unique value int this column

df['Total_Stops'].unique()
```

```python
a=df['Total_Stops'].mode() # we will replace nan value with mode as we can see 1 stop is most repetative in our data
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,np.nan:1})
```

```python
df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Hour']=df['Arrival_Time'].str.split(" ").str[0].str.split(':').str[0]
df['Minute']=df['Arrival_Time'].str.split(" ").str[0].str.split(':').str[1]
```

```python
df['Hour']=df['Hour'].astype(int)
df['Minute']=df['Minute'].astype(int)
```

```python
df.head()
```

```python
df['Duration_minute']=df['Duration'].str.split("h").str.get(1).str.split('m').str[0]
```

```python
df['Duration_hour']=df['Duration'].str.split("h").str[0]
```

```python
df.head()
```

```python
df['Duration_minute'].isnull().sum() # we can see there is one na value in minute we will fill this with name
```

```python
df['Duration_minute'] = df['Duration_minute'].replace('', np.nan)
```

```python
df['Duration_minute'] = df['Duration_minute'].astype(float).astype('Int64')
```

```python
df['Duration_minute'].fillna(round(df['Duration_minute'].mean()), inplace=True)
```
